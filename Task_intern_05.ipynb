```python
import matplotlib.pyplot as plt
import pandas as pd

# Example data
data = {'City': ['City A', 'City B', 'City C', 'City D', 'City E', 'City F', 'City G', 'City H', 'City I', 'City J'],
        'Accidents': [120, 150, 90, 200, 180, 160, 140, 130, 110, 100]}
accidents_by_cities = pd.DataFrame(data)
accidents_by_cities.set_index('City', inplace=True)

# Plotting
fig, ax = plt.subplots(figsize=(8, 5))
accidents_by_cities[:10].plot(kind='bar', ax=ax)
ax.set(title='Top 10 Cities By Number of Accidents',
       xlabel='Cities',
       ylabel='Accidents Count')
plt.xticks(rotation=45)  # Optional: Rotate city names for better readability
plt.tight_layout()  # Adjust layout to fit labels and titles
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define your data here
accidents_severity = [30, 20, 25, 25]  # Example data, replace with your actual data
labels = ['Severity 1', 'Severity 2', 'Severity 3', 'Severity 4']  # Example labels

# Create the plot
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(aspect="equal"))

# Create the pie chart
wedges, texts, autotexts = ax.pie(accidents_severity, labels=labels, autopct='%1.1f%%', pctdistance=0.85)

# Draw a white circle at the center to create a 'donut' style chart
circle = plt.Circle((0, 0), 0.5, color='white', transform=ax.transData)
ax.add_artist(circle)

# Set title
ax.set_title("Accidents by Severity", fontdict={'fontsize': 16})

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
import pandas as pd
import numpy as np

# Example data
data = {
    'Weather': ['Clear', 'Rain', 'Fog', 'Snow', 'Windy', 'Clear', 'Rain', 'Snow', 'Fog', 'Windy',
                'Clear', 'Rain', 'Snow', 'Fog', 'Windy', 'Clear', 'Rain', 'Snow', 'Fog', 'Windy'],
    'Accidents': np.random.randint(1, 100, 20)
}
weather_conditions = pd.DataFrame(data)
weather_conditions = weather_conditions.groupby('Weather').sum()  # Aggregate accident counts by weather condition
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
weather_conditions.sort_values(by='Accidents', ascending=False)[:20].plot(kind='bar', ax=ax)
ax.set(title='Weather Conditions at Time of Accident Occurrence',
       xlabel='Weather',
       ylabel='Accidents Count')
plt.xticks(rotation=45)  # Optional: Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to fit labels and titles
plt.show()
```
